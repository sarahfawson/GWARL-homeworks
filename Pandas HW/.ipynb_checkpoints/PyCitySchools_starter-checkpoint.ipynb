# PyCity Schools Analysis

### OBSERVATIONS:
* Charter schools showed significantly higher grades and passing rates in both subjects than district schools. However, more analysis will be required to glean if the effect is due to school practices or the fact that charter schools tend to serve smaller student populations per school. 

* As a whole, schools with higher budgets did not yield better test results. By contrast, schools with higher spending per student actually underperformed compared to schools with smaller budgets to allocate to their students.

* Smaller and medium sized schools dramatically out-performed large sized schools on passing math performances (89-91% passing vs 67%). We do not know if this corresponds to fewer students led by each teacher or not.

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
# rename a column so it makes more sense when merged
school_data = school_data.rename(columns={"type": "school_type"})

# Calculate the total number of schools
d_total_schools = len(set(school_data["school_name"]))

# Calculate the total budget
d_total_budget = school_data["budget"].sum()

# merge the student and schools dfs
studentschools_merged = pd.merge(school_data, student_data, on="school_name")

# count number of unique (with set) student names
d_total_students = len(set(studentschools_merged["Student ID"]))

# Calculate the average math score of students
d_avg_math = studentschools_merged["math_score"].mean()

# Calculate the average reading score of students
d_avg_reading = studentschools_merged["reading_score"].mean()

# Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2
d_passing_rate = (d_avg_math + d_avg_reading)/2

# Calculate the percentage of students with a passing math score (70 or greater)
d_percent_passing_math = ((len(studentschools_merged[studentschools_merged['math_score']>=70]))/d_total_students)*100

# Calculate the percentage of students with a passing reading score (70 or greater)
d_percent_passing_reading = ((len(studentschools_merged[studentschools_merged['reading_score']>=70]))/d_total_students)*100
```

```python
#do some number formatting so the summary table looks pretty
# d_total_budget = d_total_budget.map("${:.2f}".format)

#set up summary table and assign column names
summary_table = {"Total Schools": [d_total_schools],
                 "Total Students": [d_total_students],
                 "Total Budget": [d_total_budget],
                 "Average Math Score": [d_avg_math],
                 "Average Reading Score": [d_avg_reading],
                 "% Passing Math": [d_percent_passing_math],
                 "% Passing Reading": [d_percent_passing_reading],
                 "% Overall Passing Rate": [d_passing_rate],
                }
#turn summary table into a df and print
df_summary = pd.DataFrame(summary_table)
df_summary.head()
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
#--- you need to make a dataframe that has the calculation of values per school ---
school_stats_summary = {"School Name": [],
                         "School Type": [],
                         "Total Students": [],
                         "Total School Budget": [],
                         "Per Student Budget": [],
                         "Average Math Score": [],
                         "Average Reading Score": [],
                         "% Passing Math": [],
                         "% Passing Reading": [],
                         "% Overall Passing Rate": []
                        }
# make a list of schools to loop through
schools = list(set(studentschools_merged["school_name"]))

# loop through previous merged df to find all the summary data per school
for each_school in schools:
    school_df = studentschools_merged[studentschools_merged["school_name"]==each_school]
    
    # display the school type through a python type of v-lookup
    s_type = school_data[school_data["school_name"]==each_school]["school_type"].values[0]
    
    # find total students per school by accumulating each row (student) each time the formula loops
    s_total_students = len(school_df)

    # find the total budget per school and budget per student
    s_total_budget = school_data[school_data["school_name"]==each_school]["budget"].values[0]

    s_budget_per_student = s_total_budget/s_total_students

    # find the average math score per school
    s_avg_math = school_df["math_score"].mean()

    # find the average reading score per school
    s_avg_reading = school_df["reading_score"].mean()

    # find the % passing math and reading per school
    s_percent_passing_math = ((len(school_df[school_df["math_score"]>=70]))/s_total_students)*100
    s_percent_passing_reading = ((len(school_df[school_df["reading_score"]>=70]))/s_total_students)*100

    # find overall % passing rate per school
    s_passing_rate = (s_percent_passing_math + s_percent_passing_reading)/2
    
    # append the summary table with the data you just found
    school_stats_summary['School Name'].append(each_school)
    school_stats_summary['School Type'].append(s_type)
    school_stats_summary['Total Students'].append(s_total_students)
    school_stats_summary['Total School Budget'].append(s_total_budget)
    school_stats_summary['Per Student Budget'].append(s_budget_per_student)
    school_stats_summary['Average Math Score'].append(s_avg_math)
    school_stats_summary['Average Reading Score'].append(s_avg_reading)
    school_stats_summary['% Passing Math'].append(s_percent_passing_math)
    school_stats_summary['% Passing Reading'].append(s_percent_passing_reading)
    school_stats_summary['% Overall Passing Rate'].append(s_passing_rate)

#turn summary table into a df
school_stats_summary_df = pd.DataFrame(school_stats_summary)
school_stats_summary_df.head()

#set_index as school name
school_stats_summary_df.set_index("School Name",inplace = True)

#sort by highest passing rate and print table
school_stats_summary_highest = school_stats_summary_df.sort_values(by=["% Overall Passing Rate"],ascending=False)
school_stats_summary_highest.head()
```

## Bottom Performing Schools (By Passing Rate)

* Sort and display the five worst-performing schools

```python
#sort by lowest passing rate and print table
school_stats_summary_lowest = school_stats_summary_highest.sort_values("% Overall Passing Rate",ascending=True)
school_stats_summary_lowest.head()
```

## Math Scores by Grade

* Create a table that lists the average Math Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
# create summary table to put looped data into
school_math_summary = {"School Name": [],
                         "9th": [],
                         "10th": [],
                         "11th": [],
                         "12th": [],
                       }

# make a list of grades and schools
grades = list(set(studentschools_merged["grade"]))
schools = list(set(studentschools_merged["school_name"]))

# loop to find the average math score per grade and school
for school in schools:
    grade_school_math_df = studentschools_merged[studentschools_merged["school_name"]==school]
    school_math_summary["School Name"].append(school)

    for grade in grades:
        mean = grade_school_math_df[grade_school_math_df['grade'] == grade]['math_score'].mean()
        school_math_summary[grade].append(mean)
    
# turn summary table into a df and print
grade_math_summary = pd.DataFrame(school_math_summary)
grade_math_summary.sort_values(by=['School Name'], inplace = True)
grade_math_summary.set_index("School Name",inplace = True)

grade_math_summary.head()
```

## Reading Score by Grade 

* Perform the same operations as above for reading scores

```python
# create summary table to put looped data into
school_reading_summary = {"School Name": [],
                          "9th": [],
                          "10th": [],
                          "11th": [],
                          "12th": [],
                       }

# make a list of grades and schools
grades = list(set(studentschools_merged["grade"]))
schools = list(set(studentschools_merged["school_name"]))

# loop to find the average math score per grade and school
for school in schools:
    grade_school_reading_df = studentschools_merged[studentschools_merged["school_name"]==school]
    school_reading_summary["School Name"].append(school)

    for grade in grades:
        mean = grade_school_reading_df[grade_school_reading_df['grade'] == grade]['reading_score'].mean()
        school_reading_summary[grade].append(mean)
    
# turn summary table into a df and print
grade_reading_summary = pd.DataFrame(school_reading_summary)
grade_reading_summary.sort_values(by=['School Name'], inplace = True)
grade_reading_summary.set_index("School Name",inplace = True)

grade_reading_summary.head()
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Set bins
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
```

```python
# take columns from previous df and do a a pd.cut to apply the bins

school_reading_data = school_stats_summary_df[["Per Student Budget","Average Math Score","Average Reading Score","% Passing Math", "% Passing Reading","% Overall Passing Rate"]]
school_reading_data["Spending Ranges (Per Student)"] = pd.cut(school_stats_summary_df["Per Student Budget"], spending_bins, labels=group_names)

# then get an average across all the schools
school_reading_data = school_reading_data.groupby("Spending Ranges (Per Student)").mean()

# then remove the per student budget column after you're done analyzing it
school_reading_data.drop(columns=["Per Student Budget"],inplace=True)

# print
school_reading_data.head()
```
