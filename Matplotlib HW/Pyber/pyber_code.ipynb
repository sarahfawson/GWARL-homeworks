# Pyber Homework

In this project for my analytics class, I analyzed data on a ride sharing company called "Pyber." I use pandas and matplotlib to find and visualize data based on the type of city that Pyber rides take place in.

----
### Observed Data Trends:
1. The majority of rides for pyber are taken in urban cities, followed by suburban cities then the lowest volume of rides is in rural areas. This may suggest that pyber use is correlated to the population of an area.
2. Ride fares tend to be larger in rural areas than in urban cities. This could be due to distances traveled.
3. Of all Pyber drivers shown in this dataset, there is a larger percentage of drivers in urban areas (81%) as compared to the percentage of total rides that have been measured in urban areas (68%) as compared to the data from suburban and rural areas. This could mean that driving for pyber is a part-time job for more people in urban areas and a full time job for more people in suburban or rural areas, or it could just be an indicator of lower demand for drivers in less populated areas. We do not have driver utilization rate or hourly data to tell this.
----  

```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# import data files
city_data = pd.read_csv("data/city_data.csv")
ride_data = pd.read_csv("data/ride_data.csv")

# city_data.head()
# ride_data.head()

# Combine the data into a single dataset and print
pyber_data = ride_data.merge(city_data, on = 'city', how = 'outer')
pyber_data.head()
```

## Bubble Plot of Ride Sharing Data

```python
# list city types
city_type = pyber_data[["city","type"]].drop_duplicates("city")
```

```python
#Average Fare Per City
average_fare = pyber_data.groupby("city")["fare"].mean()
average_fare = pd.DataFrame(average_fare).reset_index()
average_fare = average_fare.rename(columns = {'fare': 'avg_fare'})

#Total Number of Rides Per City
total_rides = pyber_data.groupby("city")["ride_id"].count()
total_rides = pd.DataFrame(total_rides).reset_index()
total_rides = total_rides.rename(columns = {'ride_id': 'total_rides'})

#Total Drivers
total_drivers = pyber_data[["city", "driver_count"]].drop_duplicates("city")

#City Type
city_type = pyber_data[["city", "type"]].drop_duplicates("city")
bubble_df = pd.merge(pd.merge(pd.merge(average_fare, total_rides, on="city"), 
                             total_drivers, on="city"), city_type, on="city")
bubble_df.head()
```

```python
# Separate dataframes based on city types
urban_city = bubble_df.loc[(bubble_df["type"] == "Urban")]
suburban_city = bubble_df.loc[(bubble_df["type"] == "Suburban")]
rural_city = bubble_df.loc[(bubble_df["type"] == "Rural")]
```

```python
# Build the scatter plots for each city type
ax1 = urban_city.plot(kind='scatter',x='total_rides', y="avg_fare",
                       color="lightcoral", s=bubble_df["driver_count"]*4, label = "Urban", 
                       alpha = 0.5, edgecolor = "black", linewidths = 1.5)    
ax2 = suburban_city.plot(kind='scatter', x='total_rides', y="avg_fare", 
                          color="lightskyblue",s=bubble_df["driver_count"]*4, label = "Suburban", 
                          alpha = 0.5, edgecolor = "black", linewidths = 1.5, ax=ax1)    
ax3 = rural_city.plot(kind='scatter', x='total_rides', y="avg_fare", 
                       color="gold", s=bubble_df["driver_count"]*4, label = "Rural", 
                       alpha = 0.5, edgecolor = "black", linewidths = 1.5, ax=ax1)

# Incorporate the other graph properties
plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")

# Create a legend
plt.legend(title = 'City Types')
plt.grid(True)

# text label about the circle size
textstr = "Note: \n Circle size corresponds with driver count per city."
plt.text(42,35,textstr, fontsize=11)

plt.show()
plt.savefig("Images/pyber_ride_sharing_data.png")
```

## Total Fares by City Type

```python
# Calculate Type Percents
fares_by_city_type = pyber_data[["fare","type"]]
fares_by_city_type_df = fares_by_city_type.groupby('type').sum()

# assign pie chart properties
labels = ["Rural", "Suburban", "Urban"]
sizes = list(fares_by_city_type_df["fare"])
colors = ["gold", "lightskyblue", "lightcoral"]
explode = (0, 0, 0.1)

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=90)

plt.title("% of Total Fares by City Type")

# Save Figure
plt.savefig("Images/percent_total_fares_by_city_type.png")
```

## Total Rides by City Type

```python
# Calculate Type Percents
rides_by_city_type = pyber_data[["ride_id","type"]]
rides_by_city_type_df = rides_by_city_type.groupby('type').count()
rides_by_city_type_df.rename(columns={"ride_id": "Ride Count"},inplace=True)

# assign pie chart properties
labels = ["Rural", "Suburban", "Urban"]
sizes = list(rides_by_city_type_df["Ride Count"])
colors = ["gold", "lightskyblue", "lightcoral"]
explode = (0, 0, 0.1)

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=90)

plt.title("% of Rides by City Type")

# Save Figure
plt.savefig("Images/percent_rides_by_city_type.png")
```

## Total Drivers by City Type

```python
# Calculate Type Percents
drivers_by_city_type = pyber_data[["city","driver_count","type"]].drop_duplicates()
drivers_by_city_type_df = drivers_by_city_type.groupby("type").sum()

# assign pie chart properties
labels = ["Rural", "Suburban", "Urban"]
sizes = list(drivers_by_city_type_df["driver_count"])
colors = ["gold", "lightskyblue", "lightcoral"]
explode = (0, 0, 0.1)

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=90)

plt.title("% of Total Drivers by City Type")

# Save Figure
plt.savefig("Images/percent_drivers_by_city_type.png")
```
